```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file into a pandas DataFrame
data = pd.read_csv('score3.csv')

# Filter values above the threshold
threshold = 0.93
filtered_data = data[data['R2'] > threshold]

# Extract the filtered R2 and MAE values
r2_values = filtered_data['R2']
mae_values = filtered_data['MAE']

# Extract the combination of the first three columns
combinations = filtered_data[['Layer 1', 'Layer 2', 'Layer 3']].apply(lambda x: f"({x['Layer 1']},{x['Layer 2']},{x['Layer 3']})", axis=1)

# Plot R2 scores with combinations as x-axis labels
plt.figure(figsize=(12, 8))
plt.plot(range(len(r2_values)), r2_values, marker='o', color='purple')
plt.xlabel('Hidden Layer Combinations')
plt.ylabel('R2')
plt.title('R2 Scores for Each Combination (R2 > 0.93)')

# Set the x-axis labels as the combinations and rotate them vertically
plt.xticks(range(len(r2_values)), combinations, rotation='vertical')

# Add some spacing between the points on the x-axis
plt.subplots_adjust(bottom=0.2)

plt.tight_layout()
plt.show()

# Plot MAE scores with combinations as x-axis labels
plt.figure(figsize=(12, 8))
plt.bar(range(len(mae_values)), mae_values, color='dodgerblue')
plt.xlabel('Hidden Layer Combinations')
plt.ylabel('MAE')
plt.title('MAE Scores for Each Combination (R2 > 0.93)')

# Set the x-axis labels as the combinations and rotate them vertically
plt.xticks(range(len(mae_values)), combinations, rotation='vertical')

# Add some spacing between the points on the x-axis
plt.subplots_adjust(bottom=0.2)

plt.tight_layout()
plt.show()
```
